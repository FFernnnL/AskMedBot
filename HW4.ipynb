```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000000
x = np.zeros(N)
x[0] = 0

delta = 0.1
kT = 1

for i in range(1, N):
    x_new = x[i-1] + delta * np.random.randn()
    delta_E = (x_new**2 - x[i-1]**2) / 2
    if np.random.rand() < np.exp(-delta_E):
        x[i] = x_new
    else:
        x[i] = x[i-1]

mean_x = np.mean(x)
mean_x2 = np.mean(x**2)
mean_x3 = np.mean(x**3)
mean_x4 = np.mean(x**4)

print(f"Mean(x) = {mean_x}")
print(f"Mean(x^2) = {mean_x2}")
print(f"Mean(x^3) = {mean_x3}")
print(f"Mean(x^4) = {mean_x4}")

plt.hist(x, bins=100, density=True, alpha=0.6)
x_vals = np.linspace(-5, 5, 100)
pdf = np.exp(-x_vals**2 / 2) / np.sqrt(2 * np.pi)
plt.plot(x_vals, pdf, color='red')
plt.xlabel('x')
plt.ylabel('Density')
plt.show()
```
